# 1. ä½ç½®æå–ROI

```python
import cv2 #opencvçš„ç¼©å†™ä¸ºcv2
import matplotlib.pyplot as plt # matplotlibåº“ç”¨äºç»˜å›¾å±•ç¤º
import numpy as np   # numpyæ•°å€¼è®¡ç®—å·¥å…·åŒ…
```

```python
def cv_show(name,img):
    cv2.imshow(name,img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

img = cv2.imread('01_Picture/01_cat.jpg')
print(img.shape)
cat = img[0:200,0:200,:] # é€‰æ‹©å›¾ç‰‡æ„Ÿå…´è¶£çš„åŒºåŸŸ
cv_show('cat',cat)
```

# 2. é€šé“æå–ROI

## 2.1 åˆ†ç¦» BGR é€šé“

```python
img = cv2.imread('01_Picture/01_cat.jpg')
b,g,r = cv2.split(img)
cv_show('cat_b',b)
print('b.shape:',b.shape) # Bé€šé“ï¼Œå•é€šé“ï¼Œç°åº¦å›¾
cv_show('cat_g',g)
print('g.shape:',g.shape) # Gé€šé“ï¼Œå•é€šé“ï¼Œç°åº¦å›¾
cv_show('cat_r',r)
print('r.shape:',r.shape) # Ré€šé“ï¼Œå•é€šé“ï¼Œç°åº¦å›¾
img = cv2.merge((b,g,r))
print('img.shape:',img.shape) # 3 é€šé“ï¼Œå½©è‰²å›¾
```

## 2.2 å±•ç¤º R é€šé“

```python
# åªä¿ç•™ R
img = cv2.imread('01_Picture/01_cat.jpg')
b,g,r = cv2.split(img)
img = cv2.merge((b,g,r))
cur_img = img.copy()
cur_img[:,:,0] = 0 
cur_img[:,:,1] = 0
cv_show('R',cur_img)
```
